## Description of the Context: Bounded Knapsack Problem

You are given a set of items, each with an associated value and weight, and a knapsack with a maximum weight capacity. Each item is available a certain number of times. The goal is to determine the most valuable combination of items to place in the knapsack without exceeding its weight capacity.

**Problem Parameters:**

- **Set of Items:** There is a finite set of items, often denoted as $I = \{1, 2, \ldots, n\}$.

- **Item Properties:**
  - Each item $i$ has an associated value $v_i$ (positive integer number), representing the profit gained by including the item in the knapsack.
  - Each item $i$ has an associated weight $w_i$ (positive integer number), indicating the weight of the item.
  - Each item $i$ has an associated maximum quantity $b_i$ (a non-negative integer), specifying the maximum number of times that item can be selected.

- **Knapsack Capacity:** The knapsack has a maximum weight capacity, denoted as $W$ (a positive integer number).

**Objective:**

The objective of the bounded knapsack problem is to determine a selection of items, taking into account the quantity bounds, that maximizes the total value of the selected items while ensuring that the total weight does not exceed the knapsack's capacity.


## I. Fractional Bounded Knapsack

In the fractional version of the problem, you are allowed to break items into bits as tiny as you wish for.

**Q1.** Write a function `knapfrac(values, weights, quantities, capacity)` whose arguments are:
 - a list of values (values of items), 
 - a list of weights (weights of items), 
 - and a list of bounds (upper bounds on the number of items of each type),
 - an integer capacity (maximum weight the knapsack can hold), 
 
 and which returns the maximum value of a fractional knapsak.

```python
def knapfrac(values, weights, quantities, capacity):
    n = len(values)
    items = []
    
    # On va construire une liste (ratio, poids_max_total)
    for i in range(n):
        v = values[i]
        w = weights[i]
        b = quantities[i]
        if w <= 0 or b <= 0:
            continue  # rien à prendre ici 
        ratio = v / w
        max_weight = w * b  # poids total max si on prend b objets
        items.append((ratio, max_weight))
    
    # On trie par ratio decroissant
    items.sort(key=lambda x: x[0], reverse=True)
    
    total_value = 0.0
    remaining_capacity = capacity
    
    for ratio, max_weight in items:
        if remaining_capacity <= 0:
            break
        
        # C'est le poids que l'on peut encore prendre
        take_weight = min(remaining_capacity, max_weight)
        
        # On augmente la valeur totale
        total_value = total_value + take_weight * ratio
        
        # On met a jour la capacite restante
        remaining_capacity = remaining_capacity - take_weight
    
    return total_value


#Test 
values1 = [60, 100, 120]
weights1 = [10, 20, 30]
quantities1 = [2, 1, 3]
capacity1 = 50
    
res1 = knapfrac(values1, weights1, quantities1, capacity1)
print("Test 1:",res1)

```

```python
def knapfrac_opt(values, weights, quantities, capacity):
    n = len(values)
    items = []
    
    # Initialisons la solution (quantite de chaque item pris)
    solution = [0.0] * n
    
    # On va construire une liste (ratio, poids_max_total, indice, poids_unitaire)
    for i in range(n):
        v = values[i]
        w = weights[i]
        b = quantities[i]
        if w <= 0 or b <= 0:
            continue  # rien a prendre ici 
        ratio = v / w
        max_weight = w * b  # poids total max si on prend b objets
        items.append((ratio, max_weight, i, w))
    
    # On trie par ratio decroissant
    items.sort(key=lambda x: x[0], reverse=True)
    
    total_value = 0.0
    remaining_capacity = capacity
    
    for ratio, max_weight, idx, unit_weight in items:
        if remaining_capacity <= 0:
            break
        
        # C'est le poids que l'on peut encore prendre
        take_weight = min(remaining_capacity, max_weight)
        
        # On augmente la valeur totale
        total_value = total_value + take_weight * ratio
        
        # On calcule la quantite d'items pris (en nombre d'items)
        quantity_taken = take_weight / unit_weight
        solution[idx] = quantity_taken
        
        # On met a jour la capacite restante
        remaining_capacity = remaining_capacity - take_weight
    
    return total_value, solution
```

# II. Bounded Knapsack Dynamically

**Q0.** Provide and explain a recurrence relation which the bounded knapsack problem satisfies.

**Answer :**
On définit :   
K(i,w) = la valeur maximale qu'on peut obtenir en utilisant les i premiers items avec une capacité maximale de w. 

Pour chaque item i, on note :  
-vi: la valeur de l'item i  
-wi: le poids de l'item i  
-bi: la quantité maximale disponible de l'item i.

Pour calculer la valeur optimale avec les i premiers items et une capacité w, on teste toutes les quantités possibles k de l'item i que l'on peut prendre. Pour chaque quantité k :

On utilise un poids total de k×wi. On obtient une valeur de k×vi, et il reste une capacité de (w - k×wi) pour les i-1 premiers items. 

La quantité k est limitée par deux contraintes :
1) k ≤ bi (on ne peut pas dépasser la quantité disponible) 
2) k×wi ≤ w (on ne peut pas dépasser la capacité du sac)

On choisit ensuite la quantité k qui maximise la valeur totale.

On obtient donc la récurrence : 

K(i,w) = max { k×vi + K(i-1, w - k×wi) } pour  0 ≤ k ≤ bi **et** k×wi ≤ w  

Les cas de base:  
-K(0,w) = 0 pour tout w ≥ 0 (i.e aucun item disponible)   
-K(i,0) = 0 pour tout i ≥ 0 (i.e capacité nulle)




**Q1.** Write a function `knapdyn(values, weights, quantities, capacity)` which returns the maximum value of a bounded knapsack problem.

```python
def knapdyn(values, weights, quantities, capacity):
    n = len(values)
    
    # On cree le tableau de programmation dynamique
    # K[i][w] = valeur max avec les i premiers items et capacite w
    K = [[0 for _ in range(capacity + 1)] for _ in range(n + 1)]
    
    # On remplit le tableau K de maniere iterative
    for i in range(1, n + 1):
        v = values[i - 1]  # valeur de l'item i
        w = weights[i - 1]  # poids de l'item i
        b = quantities[i - 1]  # quantité max de l'item i
        
        for cap in range(capacity + 1):
            # si k=0 : on ne prend pas l'item i
            K[i][cap] = K[i - 1][cap]
            
            # On teste toutes les quantites possibles k de l'item i, i.e k=1...max_k
            if (w > 0):
                max_k = min(b, cap // w) 
            else:
                max_k = 0
            
            for k in range(1, max_k + 1):
                if cap >= k * w:
                    valeur = K[i - 1][cap - k * w] + k * v
                    K[i][cap] = max(K[i][cap], valeur)
    
    return K[n][capacity]


print("TEST 1: Exemple de base")
values1 = [60, 100, 120]
weights1 = [10, 20, 30]
quantities1 = [2, 1, 3]
capacity1 = 50

result1 = knapdyn(values1, weights1, quantities1, capacity1)
print("Valeur optimale: ", result1)
print()
    
print("TEST 2: Tous les items")
values2 = [10, 20, 30]
weights2 = [1, 2, 3]
quantities2 = [5, 5, 5]
capacity2 = 30

result2 = knapdyn(values2, weights2, quantities2, capacity2)
print("Valeur optimale: ", result2)
print()

print("TEST 3: Un seul item")
values3 = [100]
weights3 = [50]
quantities3 = [3]
capacity3 = 100
    
result3 = knapdyn(values3, weights3, quantities3, capacity3)
print("Valeur optimale: ", result3)
print("Attendu: 200 (2 items)")
print()
    
print("TEST 4: Capacite nulle")
values4 = [10, 20, 30]
weights4 = [1, 2, 3]
quantities4 = [5, 5, 5]
capacity4 = 0
    
result4 = knapdyn(values4, weights4, quantities4, capacity4)
print("Valeur optimale: ", result4)
print("Attendu: 0")
print()
    
print("TEST 5: Items avec quantites limitees")
values5 = [50, 60, 70]
weights5 = [10, 15, 20]
quantities5 = [3, 2, 1]
capacity5 = 45
    
result5 = knapdyn(values5, weights5, quantities5, capacity5)
print("Valeur optimale: ", result5)
print()
```

**Q2.** Enhance `knapdyn` into an efficient function `knapdynMem` which uses memoization.

```python
def knap_recursive(i, cap, values, weights, quantities, memo):
    # Cas de base
    if i == 0 or cap == 0:
        return 0

    # si deja calculé, on return 
    if (i, cap) in memo:
        return memo[(i, cap)]

    v = values[i - 1]
    w = weights[i - 1]
    b = quantities[i - 1]

    # Cas ou on ne prend pas l'item i
    max_value = knap_recursive(i - 1, cap, values, weights, quantities, memo)

    # On teste toutes les quantites possibles
    if w > 0:
        max_k = min(b, cap // w)
    else:
        max_k = 0

    for k in range(1, max_k + 1):
        valeur = knap_recursive(i - 1, cap - k * w, values, weights, quantities, memo) + k * v
        if valeur > max_value:
            max_value = valeur

    memo[(i, cap)] = max_value
    return max_value


def knapdynMem(values, weights, quantities, capacity):
    memo = {}  #dictionnaire pour les calculs deja effectue
    n = len(values)
    return knap_recursive(n, capacity, values, weights, quantities, memo)

print("TEST 1: Exemple de base")
values1 = [60, 100, 120]
weights1 = [10, 20, 30]
quantities1 = [2, 1, 3]
capacity1 = 50
    
result1 = knapdynMem(values1, weights1, quantities1, capacity1)
print("Valeur optimale: ", result1)
print()
```

```python
def knapdyn_opt(values, weights, quantities, capacity):
    n = len(values)
    
    # On cree le tableau de programmation dynamique
    # K[i][w] = valeur max avec les i premiers items et capacite w
    K = [[0 for _ in range(capacity + 1)] for _ in range(n + 1)]
    
    # On remplit le tableau K de maniere bottom-up
    for i in range(1, n + 1):
        v = values[i - 1]  # valeur de l'item i
        w = weights[i - 1]  # poids de l'item i
        b = quantities[i - 1]  # quantite max de l'item i
        
        for cap in range(capacity + 1):
            # Initialiser avec le cas ou on ne prend pas l'item i
            K[i][cap] = K[i - 1][cap]
            
            # Tester toutes les quantites possibles k de l'item i
            if (w > 0):
                max_k = min(b, cap // w) 
            else:
                max_k = 0
            
            for k in range(1, max_k + 1):
                if cap >= k * w:
                    valeur = K[i - 1][cap - k * w] + k * v
                    K[i][cap] = max(K[i][cap], valeur)
    
    # On reconstitue la solution en remontant le tableau
    solution = [0] * n
    cap = capacity
    
    for i in range(n, 0, -1):
        v = values[i - 1]
        w = weights[i - 1]
        b = quantities[i - 1]
        
        # Trouver combien d'items i ont ete pris
        if (w > 0):
            max_k = min(b, cap // w) 
        else:
            max_k = 0
        
        for k in range(max_k, -1, -1):
            if cap >= k * w:
                if K[i][cap] == K[i - 1][cap - k * w] + k * v:
                    solution[i - 1] = k
                    cap = cap - k * w
                    break
    
    return K[n][capacity], solution
```

```python
def knap_recursive_opt(i, cap, values, weights, quantities, memo, decisions):
    # Cas de base
    if i == 0 or cap == 0:
        return 0

    # si deja calculé, on return
    if (i, cap) in memo:
        return memo[(i, cap)]

    # Propriete de l'item i
    v = values[i - 1]
    w = weights[i - 1]
    b = quantities[i - 1]

    # Cas ou on ne prend pas l'item i
    max_value = knap_recursive_opt(i - 1, cap, values, weights, quantities, memo, decisions)
    best_k = 0  # quantite choisie pour l'item i

    # Nombre max de copies possibles de l'item i
    if w > 0:
        max_k = min(b, cap // w)
    else:
        max_k = 0

    # On teste k = 1..max_k
    for k in range(1, max_k + 1):
        valeur = knap_recursive_opt(i - 1, cap - k * w,
                                    values, weights, quantities, memo, decisions) + k * v
        if valeur > max_value:
            max_value = valeur
            best_k = k

    # Memoisation valeur + decision
    memo[(i, cap)] = max_value
    decisions[(i, cap)] = best_k
    return max_value


def knapdynMem_opt(values, weights, quantities, capacity):
    n = len(values)
    memo = {}
    decisions = {}

    # On calcul de la valeur optimale
    optimal_value = knap_recursive_opt(n, capacity, values, weights, quantities, memo, decisions)

    # On reconstruit la solution
    solution = [0] * n
    cap = capacity

    for i in range(n, 0, -1):
        k = decisions.get((i, cap), 0)  # 0 par defaut s'il n'y a pas de decision
        solution[i - 1] = k
        cap -= k * weights[i - 1]

    return optimal_value, solution



#Tests
values1     = [60, 100, 120]
weights1    = [10, 20, 30]
quantities1 = [1, 1, 1]
capacity1   = 50

val1, sol1 = knapdynMem_opt(values1, weights1, quantities1, capacity1)
print("Test 1 - valeur :", val1)
print("Test 1 - solution :", sol1)
print()
values2     = [10, 40, 30]
weights2    = [5, 4, 6]
quantities2 = [3, 2, 2]
capacity2   = 15

val2, sol2 = knapdynMem_opt(values2, weights2, quantities2, capacity2)
print("Test 2 - valeur :", val2)
print("Test 2 - solution :", sol2)
print()
values3     = [10, 20, 30]
weights3    = [1,  2,  3]
quantities3 = [5,  5,  5]
capacity3   = 6

val3, sol3 = knapdynMem_opt(values3, weights3, quantities3, capacity3)
print("Test 3 - valeur :", val3)
print("Test 3 - solution :", sol3)
print()
values4     = [10, 100]
weights4    = [5,  20]
quantities4 = [10, 1]
capacity4   = 1000

val4, sol4 = knapdynMem_opt(values4, weights4, quantities4, capacity4)
print("Test 4 - valeur :", val4)
print("Test 4 - solution :", sol4)
print()
values5     = [50, 60, 70]
weights5    = [5,  10, 20]
quantities5 = [0,  3,  1]   # le premier objet est indisponible
capacity5   = 40

val5, sol5 = knapdynMem_opt(values5, weights5, quantities5, capacity5)
print("Test 5 - valeur :", val5)
print("Test 5 - solution :", sol5)
print()
```

# III. Knapsack by Branch and Bound

Branch and bound is a general algorithmic technique used for solving optimization problems, including the knapsack problem. It systematically explores a search tree of potential solutions, pruning branches that cannot lead to an optimal solution.

The primary steps in the branch and bound approach are as follows:

 - **Branching:** The problem space is divided into smaller subproblems, typically through binary choices. For example, in the knapsack problem, each item can be either included (1) or excluded (0).

 - **Bounding:** At each node in the search tree, an upper bound on the optimal solution is computed. If the bound is worse than the current best solution found, that branch of the tree is pruned, as it cannot lead to an optimal solution.

 - **Backtracking:** The algorithm explores promising branches while keeping track of the best solution found so far. When it reaches a leaf node, it updates the best solution if a better one is found.

**Q1.** Which of the previous functions can be used to obtain an upper bound at each node?

**Answer:** 
La réponse est : la fonction knapfrac (ou sa version améliorée knapfrac_opt). 
En effet, elle permet de casser les items en fractions. Elle donne toujours une valeur supérieur ou égale à la solution entière optimale. C'est donc une borne supérieur. 

A chaque noeuf de l'arbre Branch and Bound : 
1) On a déjà pris certains items (décisions fixées)
2) Pour les items restants, on utilise knapfrac pour calculer la meilleur valeur possible. 
3) Si cette borne supérieure est plus petite que la solution optimale trouvée, alors on l'abandonne. 


**Q2.** Write a function `knapBnB(values, weights, quantities, capacity)`  which returns the maximum value of a bounded knapsack problem with a branch and bound approach.

```python
def bound(i, cap, val, values, weights, quantities):
#On calcule la borne superieur pour le noeud actuel. 
    
    # Cas de base: plus d'items ou plus de capacite
    if i >= len(values) or cap <= 0:
        return val
    
    # Items restants (ceux qu'on n'a pas encore consideres)
    rest_v = values[i:]
    rest_w = weights[i:]
    rest_q = quantities[i:]
    
    # On utilise knapfrac pour calculer la meilleure valeur possible (fractionnaire)
    # C'est notre borne supérieure 
    frac_val, _ = knapfrac_opt(rest_v, rest_w, rest_q, cap)
    
    return val + frac_val


def bnb(i, cap, val, best, values, weights, quantities): 
#Fonction recursive de Branch and Bound.

    n = len(values)
    
    # Cas de base : on a traité tous les objets
    if i >= n: 
        return max(best, val) #et donc on met a jour la meilleur val
    
    # Calculer la borne supérieure
    ub = bound(i, cap, val, values, weights, quantities)
    
    # Pruning
    if ub <= best:  #si borne <= meilleur solution 
        return best #inutile d’explorer plus cette branche, on retourne best.
    
    
    # Proprietes de l'item actuel
    v = values[i]
    w = weights[i]
    b = quantities[i]
    
    # Branching: on teste toutes les quantites possible de cet item
    if (w > 0):
        max_k = min(b, cap // w) 
    else:
        max_k = 0
    
    #On essaye chaque quantité k (de max_k à 0)
    #On commence par les grandes quantites pour trouver de bonnes solutions rapidement
    for k in range(max_k, -1, -1):
        new_cap = cap - k * w
        new_val = val + k * v
        
        #On continue récursivement avec l'item suivant
        best = bnb(i + 1, new_cap, new_val, best, values, weights, quantities)
    
    return best


def knapBnB(values, weights, quantities, capacity):#Retourne la valeur maximale
    
    return bnb(0, capacity, 0, 0, values, weights, quantities)

```

```python
""""
ici, on peut faire une remarque : 
Dans knapBnb, on explores les objets dans l'ordre ou ils sont donnes dans la liste (values[i] etc). 
Mais si la liste commence par des objets "nuls" (i.e dont le ratio val/poids est faible), alors notre algo
va explorer pleins de branches inutiles avant de les couper.
Donc une idee possible serait de trier tous les objets par ordre decroissant de ratio avant de lancer knapBnB. 
De cette facon, on traite les meilleurs objets en premier, et donc on augmente la probabilite de trouver une solution optimale des le debut.
Donc on aurait la version suivante : 
"""

def knapBnB_optimise(values, weights, quantities, capacity):
    # on cree une liste de dictionnaire pour pouvoir trier les objets avant de lancer notre bnb
    n=len(values)
    items = []
    for i in range(n):
        items.append({'v': values[i],'w': weights[i],'b': quantities[i],'ratio': values[i] / weights[i] if weights[i] > 0 else 0,'original_index': i })
    
    # On trie par ratio decroissant
    items.sort(key=lambda x: x['ratio'], reverse=True)
    
    # On extrait les listes triees
    sorted_v = [x['v'] for x in items]
    sorted_w = [x['w'] for x in items]
    sorted_q = [x['b'] for x in items]
    
    # on lance la recursion sur ces listes
    return bnb(0, capacity, 0, 0, sorted_v, sorted_w, sorted_q)


#Exemple de comparaison : 

# Objets nuls en premiers
values_bad = [10, 15, 20, 100, 110, 120]
weights_bad = [50, 45, 40, 10, 12, 15]
quantities_bad = [2, 2, 2, 3, 3, 3]
capacity_bad = 100

# Sans tri
time_no_sort = timeit.timeit(
    lambda: knapBnB(values_bad, weights_bad, quantities_bad, capacity_bad),
    number=10
)

# Avec tri
time_with_sort = timeit.timeit(
    lambda: knapBnB_optimise(values_bad, weights_bad, quantities_bad, capacity_bad),
    number=10
)

print("Sans tri: " + str(round(time_no_sort, 4)) + "s")
print("Avec tri: " + str(round(time_with_sort, 4)) + "s")
print("optimisation: " + str(round(time_no_sort/time_with_sort, 2)) + "x")


print("\nOn voit que l'optimisation permet d'ameliorer significativement les performances")
```

# IV. Complexity analysis

With explanations, give the respective complexities of `knapfrac`, `knapdyn`, `knapdynMem`, and `knapBnB`.

Are the previous practical comparisons consistent with these theoretic results?

**Answer:** (Write your text here, or incorporate a photo of your written answer)

Pour knapfrac, on a :  
i) On calcule le ratio pour chaque objet : O(n)    
ii) On trie les items par ratio decroissant : O(n*log(n))  
iii) On parcourt ensuite cette liste triee une fois : O(n)  
Donc la compexité en temps est : O(n*log(n))  
Et complexité en mémoire : O(n)

Pour knapdyn, on a :    
i) Une boucle sur i=1...n   
ii) Pour chaque i, une boucle sur c = 0...W     
iii) Pour chaque (i,c), on teste tous les k possibles, ou 0 ≤ k ≤ bi et k * wi ≤ w, et donc max_k. Dont le pire cas est O(B) ( ou B = max bi). Donc la complexité en temps est : O(n * W * B).
Et en memoire : O(n * W) (celle de la table)

Pour knapdynMem, on a :   
i) Chaque etat (i,c) est calculé au plus une fois,  
ii) Et pour chaque etat, on teste encore tous les k = 0,...,max_k comme dans knapdyn.     
Donc dans le pire cas, la complexité est la meme que knapdyn : O(n * W * B). Et en memoire : O(n * W) (celle de la table)

Pour knapBnB, on a :    
i) A chaque niveau, on choisit combien d'objets d'un type on prend k=0,...,bi   
ii) On pourrait donc, dans le pire cas, devoir explorer presque toutes les combinaisons possibles. Or, le nombre de noeuds dans l'arbre peut etre : ∏(bi+1) pour i = 1,...,n. Ce qui est exponentiel en n.   
iii) A chaque noeud, on calcule une borne superieur avec knapfrac (qui coute O(m*log(m)) sur les objets restants.    
Donc le pire cas : exponentiel en n.


Comparaisons que les resultats qu'on a obtenus : 

i) On avait comparé knapdynMem et knapBnB, et conclut que : "knapdynMem est souvent plus rapide et plus stable" et  
"knapBnB peut etre plus rapide avec un bon pruning". Ce qui semble coherent avec notre analyse theorique.   
En effet,  knapdynMem a une complexité pseudo-polynomiale O(n * W * B) : qui est un temps de calcul assez stable et previsible pour une taille donnee. 
Ensuite, knapBnB est exponentiel dans le pire cas : donc il peut etre tres rapide mais aussi plus lent ou instable selon les donnnee.  

ii) On avait egalement comparé KnapdynMem, knapfrac et knapdyn, et conclut que : " knapfrac est le plus rapide, et ensuite c'est knapdynMem" et "On remarque que knapdyn et knapdynMem sont plus lents mais donnent la solution optimale entiere". 
Ce qui semble aussi coherent avec les resultat. En effet, knapfrac a une complexité O(n*log(n)) alors que knapdynMem et knapdyn ont une complexité de O(n * W * B) qui depend de W et B, et donc ils font beaucoup plus de travail que le simple glouton fractionnaire (ils donnent la solution optimale entiere).  
Enfin, entre knapdynMem et knapdyn, on avait constaté que knapdynMem était plus rapide, ce qui est plutot coherent : ils ont la meme complexité theorique, mais la memoisisation peut eviter certains calculs inutiles.
